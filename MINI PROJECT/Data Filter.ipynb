```python
#  Complex Filtering and Aggregation
```

```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("fitness_tracker_data.csv")  
```

```python
# Step 1: Filter dataset where Age is between 20 and 40 and City is "Los Angeles"
filtered_df = df[(df["Age"] >= 20) & (df["Age"] <= 40) & (df["Name"] == "Aarav")]
```

```python
filtered_df
```

```python
# Step 2: Group by Gender and calculate mean & median of Age
aggregated_df = filtered_df.groupby("Gender")["Age"].agg(["mean", "median"])
```

```python
aggregated_df
```

```python
# Step 3: Sort the result in descending order by te Age column
sorted_df = aggregated_df.sort_values(by="mean", ascending=False)
```
