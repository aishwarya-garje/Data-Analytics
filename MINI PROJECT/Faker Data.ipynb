```python
from faker import Faker
```

```python
fake = Faker()
data = {
    "Name": [fake.name() for _ in range(100)],
    "Address": [fake.address() for _ in range(100)],
    "Date": [fake.date_of_birth() for _ in range(100)]
}
```

```python
df  = pd.DataFrame(data)
```

```python
df
```

```python
# 2. Create a list of 20 fake names and store them in a "Name" column
names = [fake.name() for _ in range(20)]
df_names = pd.DataFrame({"Name": names})
```

```python
# 3. Generate fake email addresses and store them in a column called "Email"
df["Email"] = [fake.email() for _ in range(100)]
```

```python
df
```

```python
# 4. Generate fake dates of birth and convert them into an "Age" column
df["Age"] = [2025 - fake.date_of_birth(minimum_age=18, maximum_age=60).year for _ in range(100)]
```

```python
df
```

```python
# 5. Generate fake product names and create a DataFrame with columns "Product_Name" and "Price"
product_data = {
    "Product_Name": [fake.word() for _ in range(100)],
    "Price": [round(fake.random_number(digits=2), 2) for _ in range(100)]
}
```

```python
df_products = pd.DataFrame(product_data)
```

```python
df_products
```
