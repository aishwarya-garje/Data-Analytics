```python
# 14. Aggregation (agg)
```

```python
# 1. Group by "Workout_Type" and calculate both the mean and median of the "Age" column
ft.groupby("Workout_Type")["Age"].agg(["mean", "median"])
```

```python
# 2. Perform multiple aggregations: find the sum, mean, and count of the "Calories_Burned" column, grouped by "Workout_Type"
ft.groupby("Workout_Type")["Calories_Burned"].agg(["sum", "mean", "count"])
```

```python
# 3. Group by "Workout_Type" and apply min() and max() functions to the "Calories_Burned" column
ft.groupby("Workout_Type")["Calories_Burned"].agg(["min", "max"])
```

```python
# 4. Group by "Workout_Type" and apply a custom aggregation function (e.g., find the range of Age within each Workout Type)
ft.groupby("Workout_Type")["Age"].agg(lambda x: x.max() - x.min())
```

```python
# 5. Group by "Gender" and calculate the average Calories_Burned, along with the total Age for each group
ft.groupby("Gender").agg({"Calories_Burned": "mean", "Age": "sum"})
```

```python
# 6. Group the data by "Workout_Type" and find both the standard deviation and variance of Calories_Burned
ft.groupby("Workout_Type")["Calories_Burned"].agg(["std", "var"])
```
